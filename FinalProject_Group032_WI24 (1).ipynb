Social media has become heavily intertwined with daily life, and it has brought forth countless benefits such as easier communication, online networking, entertainment, and more.
```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import os
```

```python
# Function used to clean social media datasets so thay can be compiled

def clean_social_media(df, social_media_platform, tik_tok = False):
    # drop all columns with nan values 
    df_clean = df.dropna(axis=1, how = "all")
    
    # drop first 4 rows that are just filler 
    df_clean = df_clean.iloc[4:]
    df_clean.columns = ["year", "num_in_million"]
    df_clean.insert(0, "platform", social_media_platform)
    
    if tik_tok == False: 
        # keeping on the year
        df_clean["year"] = df_clean["year"].apply(lambda x: int(x.split("'")[1]) + 2000)
        if social_media_platform != "Twitter": 
            # changing type of the count to int
            df_clean["num_in_million"] = df_clean["num_in_million"].str.replace(",", "").astype(int)
    
    # combining number of users per year
    df_clean = df_clean.groupby(["year", "platform"]).sum().reset_index()
    return df_clean
```

```python
# cleaning Instagram 
instagram_path = os.path.join('data', 'instagram_data.csv')
ig_df = pd.read_csv(instagram_path)

# cleaned Instagram DF
ig_df = clean_social_media(ig_df, "Instagram")
# need to now remove the data and just keep the year
ig_df = ig_df.groupby(["year", "platform"]).sum().reset_index()

ig_missing_data_df = pd.DataFrame(
    {"year":[2019, 2019, 2019, 2019, 2019, 2020, 2020, 2020], 
      "num_in_million":[1160, 1210, 1255, 1300, 1365, 1435, 1490, 1555]
    })
ig_missing_data_df = ig_missing_data_df.groupby("year")["num_in_million"].sum().reset_index()
ig_missing_data_df["platform"] = "Instagram"

ig_df = pd.concat([ig_df, ig_missing_data_df]).sort_values(by="year")
```

```python
# cleaning Facebook
facebook_path = os.path.join('data', 'facebook_data.csv')
fb_df = pd.read_csv(facebook_path)

fb_df = clean_social_media(fb_df, "Facebook")
fb_df.head()
```

```python
# cleaning Whatsapp
whatsapp_path = os.path.join('data', 'whatsapp_data.csv')
whatsapp_df = pd.read_csv(whatsapp_path)

whatsapp_df = clean_social_media(whatsapp_df, "Whatsapp")


whatsapp_missing_data_df = pd.DataFrame(
    {"year":[2018, 2018, 2018, 2018, 2019, 2019, 2019, 2019], 
      "num_in_million":[1499, 1560, 1606, 1657, 1730, 1813, 1896, 1980]
    })
whatsapp_missing_data_df = whatsapp_missing_data_df.groupby("year")["num_in_million"].sum().reset_index()
whatsapp_missing_data_df["platform"] = "Whatsapp"

whatsapp_df = pd.concat([whatsapp_missing_data_df, whatsapp_df]).sort_values(by="year")
```

```python
# cleaning TikTok
tiktok_path = os.path.join('data', 'tiktok_data.csv')
tiktok_df = pd.read_csv(tiktok_path)

tiktok_df = clean_social_media(tiktok_df, "TikTok", True)

tiktok_missing_data_df = pd.DataFrame(
    {"year":[2015, 2016, 2017, 2018, 2018, 2018, 2018, 2019, 2019, 2019, 2019], 
      "num_in_million":[0, 0, 0, 85, 133, 198, 271, 333, 381, 439, 508]
    })
tiktok_missing_data_df = tiktok_missing_data_df.groupby("year")["num_in_million"].sum().reset_index()
tiktok_missing_data_df["platform"] = "TikTok"

tiktok_df = pd.concat([tiktok_missing_data_df, tiktok_df])#.sort_values(by="year")

tiktok_df["year"] = tiktok_df["year"].apply(lambda x: int(x.strip("*")) if isinstance(x, str) else x)
```

```python
# cleaning Twitter Data 
twitter_path = os.path.join('data', 'twitter_data.csv')
twitter_df = pd.read_csv(twitter_path)

twitter_df = clean_social_media(twitter_df, "Twitter")

twitter_missing_data_df = pd.DataFrame(
    {"year":[2020, 2020, 2020, 2020], 
      "num_in_million":[807.6, 683.4, 936.2, 1289]
    })
twitter_missing_data_df = twitter_missing_data_df.groupby("year")["num_in_million"].sum().reset_index()
twitter_missing_data_df["platform"] = "Twitter"
twitter_missing_data_df["year"] = twitter_missing_data_df["year"].astype(int)

twitter_df = pd.concat([twitter_missing_data_df, twitter_df]).sort_values(by="year")
```

```python
def commonColumns(dataframes):
    # Get the columns of the first DataFrame
    common_columns = set(dataframes[0].columns)

    # Iterate through the rest of the DataFrames
    for df in dataframes[1:]:
        # Update the intersection with the columns of the current DataFrame
        common_columns = common_columns.intersection(df.columns)

    # Print the intersection of columns
    print('Common columns:')
    for x in common_columns:
        print(x)
    print()
    
def noncommonColumns(dataframes):
    # Create a dictionary to store the presence of each column across DataFrames
    column_presence = {}

    # Iterate through each DataFrame
    for i, df in enumerate(dataframes):
        # Update the presence of each column in the dictionary
        for column in df.columns:
            if column not in column_presence:
                column_presence[column] = [i]
            else:
                column_presence[column].append(i)

    # Print columns not present in all DataFrames
    print("Columns not present in all DataFrames:")
    for column, df_indices in column_presence.items():
        if len(df_indices) < len(dataframes):
            print(f"Column '{column}' is present in DataFrame(s): {', '.join(map(lambda x: str(x + 1), df_indices))}")
```

```python
def visualization_score(df, col, year):
    """
    Visualize the scores of countries based on the specified column for the top 10, middle 10, and lowest 10 countries.

    Parameters:
    - df (pandas.DataFrame): DataFrame containing the data.
    - col (str): Name of the column for which scores are visualized.
    - year (int): Year for which the visualization is performed.
    """

    top_10 = df.sort_values(by=col, ascending=False).iloc[:10]
    lowest_10 = df.sort_values(by=col).iloc[:10] 
    num_countries = len(df)
    middle_index = num_countries // 2
    middle_10 = df.sort_values(by=col).iloc[middle_index - 5:middle_index + 5]
    
    
    top_10_description = top_10[col].describe().reset_index()
    top_10_description.columns = ['Statistic', 'Value']  # Rename columns for clarity
    
    print("Top 10 Description:")
    print(top_10_description.to_string(index=False))
    print()  
    
    sns.set(font_scale=1, style="white")
    plt.figure(figsize=(16, 10))  
    
    # Plot the top 10
    plt.subplot(3, 1, 1)  
    sns.barplot(data=top_10, y='country', x=col)
    plt.title(f"10 Countries with the Highest {col} in {year}")
    
    plt.tight_layout()  
    plt.show()
    
    # Print description of the lowest 10 after the first bar plot
    middle_10_description = middle_10[col].describe().reset_index()
    middle_10_description.columns = ['Statistic', 'Value']  # Rename columns for clarity
    
    print("Middle 10 Description:")
    print(middle_10_description.to_string(index=False))
    print() 
    
    # Plot the middle 10
    plt.figure(figsize=(16, 10))  
    plt.subplot(3, 1, 2)  
    sns.barplot(data=middle_10, y='country', x=col)
    plt.title(f"10 Countries with the Middle {col} Scores in {year}")
    
    plt.tight_layout()  
    plt.show()  

    # Print description of the lowest 10 after the first bar plot
    lowest_10_description = lowest_10[col].describe().reset_index()
    lowest_10_description.columns = ['Statistic', 'Value']  # Rename columns for clarity
    
    print("Lowest 10 Description:")
    print(lowest_10_description.to_string(index=False))
    print() 
    
    # Plot the lowest 10
    plt.figure(figsize=(16, 10))  
    plt.subplot(3, 1, 3)  
    sns.barplot(data=lowest_10, y='country', x=col)
    plt.title(f"10 Countries with the Lowest {col} Scores in {year}")
    
    plt.tight_layout()  
    plt.show()  
```

```python
def overall_mean_score_bar(df, col):
    """
    Plot bar charts displaying the mean scores of the specified column for top 10, middle 10, and lowest 10 countries.

    Parameters:
    - df (pandas.DataFrame): DataFrame containing the data.
    - col (str): Name of the column for which mean scores are calculated.
    """
    df_mean = df.groupby("country")[col].mean().reset_index()
    plt.figure(figsize=(16, 20))  # Adjust figure size

    sns.set(font_scale=2.5, style="white")

    top_10 = df_mean.sort_values(by=col, ascending=False).iloc[:10]
    lowest_10 = df_mean.sort_values(by=col, ascending=True).iloc[:10]
    
    num_countries = len(df_mean)
    middle_index = num_countries // 2
    middle_10 = df_mean.sort_values(by=col).iloc[middle_index - 5:middle_index + 5]
    
    # Create subplots
    plt.subplot(3, 1, 1)  # Subplot 1
    sns.barplot(data=top_10, y='country', x=col)
    plt.title(f"Mean of {col} for Top 10 Countries over the Years (2015-2020)")
    
    plt.subplot(3, 1, 2)  # Subplot 2
    sns.barplot(data=middle_10, y='country', x=col)
    plt.title(f"Mean of {col} for Middle 10 Countries over the Years (2015-2020)")

    plt.subplot(3, 1, 3)  # Subplot 3
    sns.barplot(data=lowest_10, y='country', x=col)
    plt.title(f"Mean of {col} for Lowest 10 Countries over the Years (2015-2020)")
    
    plt.tight_layout() 
    plt.show()
```

### Why did we choose regression? 
*<b>To determine if there was significant correlation between social media users and happiness, we performed a regression analysis. This was done separately for each of the three tiers to allow for more significant correlations and comparison.</b>*

```python
def retrieve_three_tiers(df, col):
    """
    Retrieve data for countries that are part of the 3 different tiers based on the specified column.
    Parameters:
    - df (pandas.DataFrame): DataFrame containing the data.
    - col (str): Name of the column used for tier classification.
    """

    # Calculate mean for each country
    df_mean = df.groupby("country")[col].mean().reset_index()

    # Extract top 10 countries
    top_10_countries = df_mean.sort_values(by=col, ascending=False).iloc[:10]['country']
    
    # Extract bottom 10 countries
    bottom_10_countries = df_mean.sort_values(by=col).iloc[:10]['country']

    # Calculate middle index
    num_countries = len(df_mean)
    middle_index = num_countries // 2
    
    # Extract middle 10 countries
    middle_10_countries = df_mean.sort_values(by=col).iloc[middle_index - 5:middle_index + 5]['country']

    # Retrieve data for each tier
    top_10_df = df[df['country'].isin(top_10_countries)]
    bottom_10_df = df[df['country'].isin(bottom_10_countries)]
    middle_10_df = df[df['country'].isin(middle_10_countries)]

    return top_10_df, middle_10_df, bottom_10_df
```

```python
def run_regression(col_for_y, x_params, platform = False): 
    """
    Function that runs linear regression. 
    Parameters:
        - col_for_y (str): Name of the column representing the dependent variable.
        - x_params (list): List of parameter names to be used as independent variables.
        - platform (bool, optional): Whether to include platform as a feature. Defaults to False.
    """
    
    top_10_df, middle_10_df, bottom_10_df  = retrieve_three_tiers(hap, col_for_y)
    
    tiers = []
    weights = {}
    for param in x_params:
        weights[param] = []
    mse_values = []
    
    for df in [(top_10_df, "top"), (middle_10_df,"middle"), (bottom_10_df, "bottom") ]:
        temp = df[0].merge(all_media_df)
        if platform == True: 
            temp_encoded = pd.get_dummies(temp, columns=["platform"])
            X = temp_encoded[x_params]
        else: 
            X = temp[["year", "num_in_million"]]

        y = temp["happiness score"]
        reg = LinearRegression().fit(X, y)
        
        y_pred = reg.predict(X)
        mse = mean_squared_error(y, y_pred)
        mse_values.append(mse)
        tiers.append(df[1])
        for param, coef in zip(X.columns, reg.coef_):
            weights[param].append(coef)
            
    regression_df = pd.DataFrame({
        'Tier': tiers,
        'MSE': mse_values
    })
    # Add weights to the DataFrame
    for param in x_params:
        regression_df[param + '_weight'] = weights[param]

    return regression_df
```
